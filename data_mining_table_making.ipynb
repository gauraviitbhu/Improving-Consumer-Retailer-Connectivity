```python
#Transaction for mobile phones and Camera accessories

import numpy as np
import pandas as pd
from random import *
seed(10)
final_list=[]


#camera possibilities
camera_possibilities=["Camera","Focus-lens","SD-card","Camera-holder","Auto-focus-mount-adapter" , "Tripod-stand" , "Photographic-lighting"]
camera_possibilities_weights=[0.6 , 0.2 , 0.4 , 0.4 , 0.1 , 0.1, 0.07]

def makecameratransactions(camera_possibilities,k,camera_possibilities_weights,final_list):
    for i in range(50):
        final_list.append(list(set(choices(camera_possibilities,k=4+k,weights = camera_possibilities_weights))))
    for i in range(100):
        final_list.append(list(set(choices(camera_possibilities,k=2+k,weights = camera_possibilities_weights))))
    for i in range(70):
        final_list.append(list(set(choices(camera_possibilities,k=3+k,weights = camera_possibilities_weights))))
    return final_list
final_list = makecameratransactions(camera_possibilities,len(camera_possibilities)-6,camera_possibilities_weights,final_list)


#mobile-phone possibilities
mobile_possibilities = ["Mobile-phone" , "Screen-Protector" , "Sim" , "SD-card" , "Cover-Case" , "Earphones" , "Power-Bank"]
mobile_possibilities_weights = [0.7 , 0.4 , 0.5 , 0.4 , 0.2 , 0.3, 0.07]
def makemobiletransactions(mobile_possibilities,k,mobile_possibilities_weights,final_list):
    for i in range(50):
        final_list.append(list(set(choices(mobile_possibilities,k=4+k,weights = mobile_possibilities_weights))))
    for i in range(100):
        final_list.append(list(set(choices(mobile_possibilities,k=2+k,weights = mobile_possibilities_weights))))
    for i in range(70):
        final_list.append(list(set(choices(mobile_possibilities,k=3+k,weights = mobile_possibilities_weights))))
    return final_list
final_list = makemobiletransactions(mobile_possibilities,0,mobile_possibilities_weights,final_list)


#dataframe making
n=len(final_list)
l=[]
customerid=[]

def mobilenumber(row):
    return ("70237110"+str(100-int(row[5:])))

for i in range(1,41):
    if i//10 > 0:
        customerid.append("CUST0"+str(i))
    elif i>=0:
        customerid.append("CUST00"+str(i))

for i in range(1,n+1):
    if i//1000 > 0:
        l.append("TI"+str(i))
    elif i//100 > 0:    
        l.append("TI0"+str(i))
    elif i//10 > 0:
        l.append("TI00"+str(i))
    elif i>=0:
        l.append("TI000"+str(i))

df = pd.DataFrame({"TransactionID":l})
df['Transaction']=final_list
df['CustomerID']=choices(customerid,k=n)
df['Mobile_number'] = df['CustomerID'].apply(mobilenumber)
print(df)

#df.to_csv("data_mining-data.csv")
```

```python
# camera selling and costing price and profit   
# Note -: including the gst(taking it to be 12%) in bill will at last be paid by the customer and thus our profit at
# last will be, profit = selling price - cost
len(camera_possibilities)
camera_possibilities

camera_selling_prices = [50000,15000,1250,2500,14700,1750,3600]
weight_for_costing_camera = [0.7,0.64,0.95,0.75,0.85,0.67,0.75]
#camera_costing_prices = [(50000*0.7),(150000*0.64),(1250*0.95),(2500*0.75),(14700*0.85),(1750*0.67),(3600*0.0.75)]
camera_costing_prices = [camera_selling_prices[i]*weight_for_costing_camera[i] for i in range(len(camera_selling_prices))]
camera_profit = [camera_selling_prices[i] - camera_costing_prices[i] for i in range(len(camera_costing_prices))]
```

```python
#mobile sellig and costing price and profit
print(len(mobile_possibilities))
mobile_possibilities


mobile_possibilities_for_price_table = ['Mobile-phone','Screen-Protector','Sim','Cover-Case','Earphones','Power-Bank']
mobile_selling_prices = [45000,300,150,1100,2500,2500]
weight_for_costing_mobiles = [0.90,0.3,0.80,0.4,0.8,0.5]
mobile_costing_prices = [mobile_selling_prices[i]*weight_for_costing_mobiles[i] for i in range(len(mobile_selling_prices))]
mobile_profit = [mobile_selling_prices[i]-mobile_costing_prices[i] for i in range(len(mobile_possibilities_for_price_table))]
```

```python
# making the product description table

products_name = mobile_possibilities_for_price_table + camera_possibilities
products_selling_prices = mobile_selling_prices + camera_selling_prices
products_costing_prices = mobile_costing_prices + camera_costing_prices
products_profit = mobile_profit + camera_profit

products = pd.DataFrame({"Product Name":products_name})
products["Costing Price"] = products_costing_prices
products["Selling Price"] = products_selling_prices
products["Profit"] = products_profit
products
```

```python
#adding names to the df
def addname(row):
    return names[int(row[-2:])-1]

df["Customer Name"] = df["CustomerID"].apply(addname)
df
```

```python
#adding age to the df
def giveage(row):
    temp = int(row[-2:])
    if(temp>30):
        return int(((int(temp*1.45))%10)+50)
    elif temp<10:
        return int(((int(temp*0.3))%10)+10)
    else:
        return int(((int(temp*0.37))%20)+20)
    
df["Age"] = df["CustomerID"].apply(giveage)
df["Age"].nunique()
```

```python
#checking unque values count
df.groupby("Age")["CustomerID"].nunique()
```

```python
# adding the bill amount (with and without gst 12%)

def addbill_without_gst(row):
    billvalue = 0
    for i in row:
        billvalue = billvalue + int(products.loc[products["Product Name"] == i]["Selling Price"])
    return billvalue


def addbill_with_gst(row):
    return (row*1.12)

df["Billing Without GST"] = df["Transaction"].apply(addbill_without_gst)
df["Billing With GST"] = df["Billing Without GST"].apply(addbill_with_gst)
```
